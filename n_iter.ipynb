```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('font',family='Times New Roman')
```

```python
# 这里画的是所有数据集准确曲线

type = "AGA"
Google = pd.read_csv(type+"_Google_result.csv")
Amazon = pd.read_csv(type+"_Amazon_result.csv")
Apple = pd.read_csv(type+"_Apple_result.csv")
Microsoft = pd.read_csv(type+"_Microsoft_result.csv")
Tesla = pd.read_csv(type+"_Tesla_result.csv")

# 用图画出收敛
fig = plt.figure(figsize=(10, 8), dpi=600)
# plt.title("target")
plt.xlabel("Generations",size=24)
plt.ylabel("Fitness(Accuracy)",size=24)
plt.xticks(size=24)
plt.yticks(size=24)
plt.plot(Google["Unnamed: 0"],Google["0"],'-.',label="Google",linewidth = 3)
plt.plot(Amazon["Unnamed: 0"],Amazon["0"],'-.',label="Amazon",linewidth = 3)
plt.plot(Apple["Unnamed: 0"],Apple["0"],'-.c',label="Apple",linewidth = 3)
plt.plot(Microsoft["Unnamed: 0"],Microsoft["0"],'-.',label="Microsoft",linewidth = 3)
plt.plot(Tesla["Unnamed: 0"],Tesla["0"],'-.k',label="Tesla",linewidth = 3)

plt.legend(fontsize=20)
plt.show()

# 保存图像
fig.savefig("iterarion.pdf", dpi=600, format="pdf")
```
