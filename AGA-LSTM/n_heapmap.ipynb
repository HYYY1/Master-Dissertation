```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./AGA_Apple_pop.csv")
df = df.drop(columns=["Unnamed: 0"])
```

```python
# df.drop(df.head(67).index,inplace=True)
df
```

```python
p = df.iloc[:,0:6].astype(int)
p["dropout"] = df.iloc[:,6]
p["epochs"] = df.iloc[:,7]
p["acc"] = df.iloc[:,8]
p.rename(columns={"0":"nol","1":"nod","2":"nofl","3":"nosl","4":"nofd","5":"nosd","6":"dropout","7":"epochs"},inplace=True)
p
```

```python
p = p.sort_values(by='acc',ascending=False)
p = p.iloc[:50]
```

```python
des50 = p.corr()
heatmap_50 = plt.figure(figsize=(8,5),dpi=600)
sns.heatmap(des50,square=True,linewidths=0.1,annot=True,fmt='.2f',annot_kws={"fontsize":12})
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
heatmap_50.savefig("heatmap-50.pdf",dpi=600)
```

```python
p.describe().to_excel("n_50describe.xlsx")
```
